```python
import jieba as jb
from pprint import pprint
from collections import defaultdict
from gensim import corpora
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
with open('/home/cakgod/myfile/myCode/data/nlp数据/categorised-news-dataset-from-fudan-university/Fudan/train/C3-Art/C3-Art0001.txt', encoding='gbk') as f:
    data = [i.strip() for i in f.readlines()]
with open('/home/cakgod/myfile/myCode/data/nlp数据/中文停用词/哈工大停用词.txt') as f:
    stop_words = [i.strip() for i in f.readlines()]
```

```python
# 分词
cut_data = [[i for i in jb.cut(j) if i not in stop_words and i != ' ' and len(i) > 1] for j in data]
print(cut_data)
```

```python
# 生成词向量
dictionary = corpora.Dictionary(cut_data)
print(dictionary)
print(dictionary.token2id)
```

```python
# 文档词频矩阵
vectorizer = CountVectorizer()
connect_data = [' '.join(i) for i in cut_data]
doc_term_matrix = vectorizer.fit_transform(connect_data)
matrix_features = vectorizer.get_feature_names()
matrix_dict = sorted(vectorizer.vocabulary_.items(), key=lambda item: item[1])
print(doc_term_matrix)
print('每个词的编号：', matrix_dict)
print('涉及的特征：', ','.join(matrix_features))
print('涉及的特征数量：', len(matrix_features))
```

```python
# LDA
lda_model = LatentDirichletAllocation(n_components=2, random_state=888)
lda_model.fit(doc_term_matrix)

def output_topic_words(lda_model, matrix_features):
    n_top_words = 15
    # lda_model.components_给每个特征一个权重值
    for topic_idx,topic in enumerate(lda_model.components_):
        print("Topic #%d:"%topic_idx)
        # list里三个冒号，第一个冒号起反序的作用
        print(" ".join([matrix_features[i] for i in topic.argsort()[:-n_top_words-1:-1]]))

print('关键词提取')
output_topic_words(lda_model,matrix_features)

# 最小困惑度所对应的Topic就是最优的主题数 
print(u'\n困惑度：',lda_model.perplexity(doc_term_matrix,sub_sampling = False))  # doc_term_matrix 特征矩阵
```
