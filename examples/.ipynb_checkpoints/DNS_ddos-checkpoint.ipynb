# Similarity 

```python
import networkx as nx
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
plt.xkcd()
```

```python
ip_dict = {}
badip_dict = {}
# 相似度
N = 0.5
# 黑客团伙IP最少个数
M = 3
# 黑客IP攻击目标最小个数
R = 2
```

```python
# jarccard系数
def get_len(d1, d2):
    ds1 = set()
    for d in d1.keys():
        ds1.add(d)
    ds2 = set()
    for d in d2.keys():
        ds2.add(d)
    return len(ds1 & ds2) / len(ds1 | ds2)
```

```python
# read data
with open('/content/drive/My Drive/mine/dataset/DNS_ddos.txt') as f:
  for line in f:
      (ip,domain) = line.split("\t")
      if not ip == '0.0.0.0':
          if not ip in ip_dict.keys():
              ip_dict[ip] = {}
          ip_dict[ip][domain]=1
for ip in ip_dict.keys():
  if len(ip_dict[ip]) >= R:
      badip_dict[ip] = 1
```

```python
print(ip_dict)
print(badip_dict)
```

```python
print(len(ip_dict))
```

```python
G = nx.Graph()

# 检测黑团伙，高大上点叫dns ddos慢速攻击检测
for ip1 in ip_dict.keys():
  for ip2 in ip_dict.keys():
    if not ip1 == ip2 :
      weight = get_len(ip_dict[ip1], ip_dict[ip2])
      if (weight >= N) and (ip1 in badip_dict.keys()) and (ip2 in badip_dict.keys()):
        #点不存在会自动添加
        G.add_edge(ip1, ip2, weight=weight)
```

```python
print(len(list(G.nodes())))
```

```python
nx.draw(G)
```

```python
# connected component
component_list = list(nx.connected_components(G))
print('Components count: {}'.format(len(component_list)))
component_info = {i:comp for i,comp in enumerate(component_list)}
for k, v in component_info.items():
    print('Component {} with size {} has nodes: {}'.format(k, len(v), v))
```

```python
G0 = nx.subgraph(G, component_list[0])
figsize(10,8)
nx.draw(G0)
```
