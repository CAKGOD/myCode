```python
import networkx as nx
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
G = nx.read_yaml('../model/github_user_graph.yaml')
```

```python
new_G = G.to_undirected()
```

```python
print(new_G.nodes())
```

```python
users_without_relation = [i for i in new_G.nodes() if ]
```

```python
components = list(nx.connected_components(new_G))
print(f'Components count: {len(components)}')
component_sizes = {i: len(comp) for i, comp in enumerate(components)}
for k, v in component_sizes.items():
    print(f'Component {k} has size: {v}')
component_values = list(component_sizes.values())

figsize(8,6)
plt.hist(component_values)
plt.title('component size distribution')
plt.xlabel('component size')
plt.ylabel('Frequency')
plt.grid(True,linestyle = '--')
plt.show()
```

```python
sub_G = G.subgraph(components[2])
#print(sub_G.nodes['SallySoul'])
nx.draw(sub_G)
```

```python
temp = []
for i in G.nodes():
    if G.nodes[i]['followers_num'] > 1000:
        temp.append(i)
G_1 = G.subgraph(temp)

figsize(15, 15)
nx.draw(G_1)
```

```python
# degree distribution

in_degree_values = [G.nodes[i]['followers_num'] for i in G.nodes()]

figsize(24,6)

plt.subplot(1,3,1)
plt.title('Degree distribution')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.hist(in_degree_values)
plt.grid(True,linestyle = '--')

plt.subplot(1,3,2)
plt.title('Degree distribution (log scale)')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.hist(in_degree_values,log = True)
plt.grid(True,linestyle = '--')

def ecdf(array):
    return (np.sort(array), np.arange(1, len(array) + 1) / len(array))
x,y = ecdf(in_degree_values)
plt.subplot(1,3,3)
plt.title('Degree ecdf')
plt.xlabel('Degree')
plt.ylabel('Percentile')
plt.plot(x,y,linewidth = 3)
plt.grid(True,linestyle = '--')

plt.show()

print('Mean Degree: {}, Assortativity: {}'.format(round(np.mean(in_degree_values),2),round(nx.degree_assortativity_coefficient(G),2)))

degree_stats = pd.DataFrame(zip([1,10,25,50,75,90,99,100],np.percentile(in_degree_values,[1,10,25,50,75,90,99,100])),columns = ['Percentile','Value'])
print(degree_stats)
```
