```python
from xml.etree.ElementTree import parse
doc = parse('/home/cakgod/code/data/country/country_city.xml')
```

```python
country_state_city_dict = {}

root = doc.getroot()
for i in root:
    country_name = i.attrib['Name']
    country_state_city_dict[country_name] = {}
    try:
        for j in i:            
            if j.attrib == {}:
                state_name = 'None'
            else:
                state_name = j.attrib['Name']
            country_state_city_dict[country_name][state_name] = []
            
            try:
                for k in j:
                    city_name = k.attrib['Name']
                    country_state_city_dict[country_name][state_name].append(city_name)
            except:
                continue
            
    except:
        state_name = 'None'
        country_state_city_dict[country_name][state_name] = []
        print('There are not any states')
        continue
```

```python
country_name = [' '.join([j.lower() for j in i.split()]) for i in list(country_state_city_dict.keys())] + ['中国'， ]
print(len(country_name))
print(country_name)
```

```python
state_name_temp = []
for k, v in country_state_city_dict.items():
    if v == {}:
        continue
    state_name_temp += list(v.keys())
state_name = [' '.join(i.split()).lower() for i in list(set(state_name_temp))]
print(state_name)
```

```python
city_name_temp = []
for k, v in country_state_city_dict.items():
    if v == {}:
        continue
    for k1, v1 in v.items():
        city_name_temp += v1
city_name = [' '.join(i.split()).lower() for i in list(set(city_name_temp))]
print(city_name)
```
